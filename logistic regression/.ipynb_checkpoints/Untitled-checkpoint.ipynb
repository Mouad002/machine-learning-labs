```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import Normalizer

%matplotlib inline
```

```python
# Binary classification: 0 vs 1
mask = (digits.target == 0) | (digits.target == 1)
X = digits.images[mask].reshape(-1, 64)  # Flatten 8x8 images
y = digits.target[mask]
```

```python
X
```

```python
y
```

```python
scaler= Normalizer().fit(X) # the scaler is fitted to the  dataset
X = scaler.transform(X) # the scaler is applied to the dataset
```

```python
X
```

```python
y_temp = y.reshape(-1, 1)

# Combine along axis=1 (columns)
data = np.hstack((X, y_temp))
```

```python
log = LogisticRegression()
log.fit(X, y)
```

```python
y_pred = log.predict(X)
```

```python
y_pred
```

```python
plt.plot(y[:])
plt.plot(y_pred[:], 'o')
plt.show()
```

```python
print(log.coef_, log.intercept_)
```

```python
log.coef_.shape
log.intercept_.
```

```python
s = log.score(X, y)
s
```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

# Reduce to 2D for visualization
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# Train logistic regression on the 2D data
log = LogisticRegression()
log.fit(X_2d, y)

# Create mesh grid to plot decision boundary
x_min, x_max = X_2d[:, 0].min(), X_2d[:, 0].max()
y_min, y_max = X_2d[:, 1].min(), X_2d[:, 1].max()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

# Predict on the grid
Z = log.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.3, cmap='bwr')
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, cmap='bwr', edgecolor='k')
plt.title("Logistic Regression Decision Boundary (PCA-reduced)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, y_pred)
print(cm)
```

```python
# Plot confusion matrix
import seaborn as sns
import pandas as pd
# confusion matrix sns heatmap 
## https://www.kaggle.com/agungor2/various-confusion-matrix-plots
ax = plt.axes()
df_cm = cm
sns.heatmap(df_cm, annot=True, annot_kws={"size": 30}, fmt='d',cmap="Blues", ax = ax )
ax.set_title('Confusion Matrix')
plt.show()
```
