```python
from sklearn.datasets import load_breast_cancer
```

```python
ds = load_breast_cancer()
```

```python
ds
```

```python
column_names = ds['feature_names']

training_data = ds['data'][:-1]
target = ds['target'][:-1]

testing_item = ds['data'][-1:]
testing_target = ds['target'][-1:]

length =  testing_item.shape[0]
```

```python
print(training_data.shape)
print(testing_item.shape)
print(testing_target.shape)
print(length)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy', max_depth=3)
tree.fit(training_data, target)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))
plot_tree(tree, feature_names=column_names, class_names=ds['target_names'], filled=True)
plt.show()
```

```python
prediction = tree.predict(testing_item)
print("Prediction:", prediction[0])
print("Actual value:", testing_target[0])
```

```python
# trying with max_depth 1

tree1 = DecisionTreeClassifier(criterion='entropy', max_depth=1)
tree1.fit(training_data, target)

plt.figure(figsize=(15, 8))
plot_tree(tree1, feature_names=column_names, class_names=ds['target_names'], filled=True)
plt.show()

prediction1 = tree1.predict(testing_item)
print("Prediction:", prediction1[0])
print("Actual value:", testing_target)
```
